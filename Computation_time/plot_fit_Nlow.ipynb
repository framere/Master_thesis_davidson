```julia
using Plots
using DelimitedFiles
```

```julia
data = readdlm("data_iterations_Nlow.txt")

# Extract the first column (iteration number) and second column (Rnorm)
Nlow = data[:, 1]
iterations = data[:, 2]
time = data[:, 3]

# Create a scatter plot
scatter(Nlow, iterations, label="Iterations", title="Iterations vs N_low", xlabel="Nlow", ylabel="Iterations", legend=:topleft)
```

```julia
scatter(Nlow, time, label="Time", title="Time vs N_low", xlabel="Nlow", ylabel="Time (s)", legend=:topleft)
```
