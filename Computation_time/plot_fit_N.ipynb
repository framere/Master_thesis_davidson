```julia
using Plots
using DelimitedFiles
```

```julia
data = readdlm("data_iterations.txt")

# Extract the first column (iteration number) and second column (Rnorm)
N = data[:, 1]
iterations = data[:, 2]
time = data[:, 3]

# Create a scatter plot
scatter(N, iterations, label="Iterations", title="Iterations vs N", xlabel="N", ylabel="Iterations", legend=:topleft)
```

```julia
scatter(N, time, label="Time", title="Time vs N", xlabel="N", ylabel="Time (s)", legend=:topleft)
```

```julia
using Polynomials

# Fit a squared function to the data
p = fit(N, time, 2)  # 2 is the degree of the polynomial
# Generate smooth x values for the fit curve
N_fit = range(minimum(N), stop=maximum(N), length=500)
time_fit = p.(N_fit)

# Plot original data and fit
scatter(N, time, label="Time", title="Time vs N", xlabel="N", ylabel="Time (s)", legend=:topleft)
plot!(N_fit, time_fit, label="Quadratic Fit", lw=2)
```
